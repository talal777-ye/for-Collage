```python
import random

# Private inputs
Ahmed = 120
Ali = 50
Rashida = 50

print("Private Values:")
print("Ahmed:", Ahmed)
print("Ali:", Ali)
print("Rashida:", Rashida)

def split_into_shares(secret):
    """Split a value into 3 random shares such that s = a + b + c"""
    a = random.randint(1, 50)
    b = random.randint(1, 50)
    c = secret - (a + b)
    return [a, b, c]

# Step 1: Everyone splits their private number
Ahmed_shares = split_into_shares(Ahmed)
Ali_shares = split_into_shares(Ali)
Rashida_shares = split_into_shares(Rashida)

# Step 2: Distribute shares (simulate by grouping)
Ahmed_recv = [Ahmed_shares[0], Ali_shares[0], Rashida_shares[0]]
Ali_recv = [Ahmed_shares[1], Ali_shares[1], Rashida_shares[1]]
Rashida_recv = [Ahmed_shares[2], Ali_shares[2], Rashida_shares[2]]

# Step 3: Each participant sums their received shares
Ahmed_sum = sum(Ahmed_recv)
Ali_sum = sum(Ali_recv)
Rashida_sum = sum(Rashida_recv)

# Step 4: Global sum
Global_sum = Ahmed_sum + Ali_sum + Rashida_sum

print("\nComputed Global Sum =", Global_sum)
print("Expected Sum =", Ahmed + Ali + Rashida)
```

```python
import random
from sympy import symbols, Eq, solve

# Parameters
secret = 123
k = 2  # threshold
n = 3  # number of shares
prime = 1613  # large prime

# Generate random polynomial coefficients: a0 = secret
a0 = secret
a1 = random.randint(1, prime-1)  # random coefficient

print("Polynomial: f(x) = a0 + a1*x")
print("a0 (secret) =", a0)
print("a1 =", a1)

def f(x):
    return (a0 + a1*x) % prime

# Generate n shares
shares = [(i, f(i)) for i in range(1, n+1)]
print("\nGenerated Shares:")
for s in shares:
    print(s)

# Reconstruct secret using 2 shares
(x1, y1), (x2, y2) = shares[0], shares[1]

# Lagrange interpolation
def reconstruct(x1, y1, x2, y2):
    num = (x2 * y1 - x1 * y2) % prime
    den = (x2 - x1) % prime
    inv_den = pow(den, -1, prime)
    return (num * inv_den) % prime

recovered_secret = reconstruct(x1, y1, x2, y2)

print("\nRecovered Secret =", recovered_secret)
print("Original Secret =", secret)
```
